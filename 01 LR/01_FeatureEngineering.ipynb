```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from talib import abstract

plt.style.use("seaborn")
```

```python
df = pd.read_csv("bid_ask.csv", parse_dates = ["time"], index_col = "time")
```

```python
df
```

```python
df['Ask_Dir'] = np.where(df['ask'].shift(-1) > df.ask, 1, 0)


df['SMA'] = abstract.SMA(df['ask'], timeperiod=12)
df['SMA_Dir'] = np.where(df['SMA'].shift(-1) > df.SMA, 1, 0)


df['RSI'] = abstract.RSI(df['ask'], timeperiod=12)
df['RSI_Dir'] = np.where(df['RSI'].shift(-1) > df.RSI, 1, 0)

df['fastk'], df['fastd'] = abstract.STOCHRSI(df["ask"], timeperiod=14, fastk_period=5, fastd_period=3, fastd_matype=0)
df['fastk_Dir'] = np.where(df['fastk'].shift(-1) > df.fastk, 1, 0)
df['fastd_Dir'] = np.where(df['fastd'].shift(-1) > df.fastd, 1, 0)

df
```
